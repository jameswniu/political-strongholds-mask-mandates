# Data Exploration & Cleanup

Sources:

"U.S. State and Territorial Public Mask Mandates From April 10, 2020 through August 15, 2021 by County by Day" - 
https://data.cdc.gov/Policy-Surveillance/U-S-State-and-Territorial-Public-Mask-Mandates-Fro/62d6-pm5i


US Accidents (updated): A Countrywide Traffic Accident Dataset (2016 - 2020) - 
https://www.kaggle.com/sobhanmoosavi/us-accidents?select=US_Accidents_Dec20.csv

The goal of this data exploration and cleanup process is to condense the two datasets above into two clean and concise datasets for the states of GA and NY. This will facilitate a comparison between the two states, and it will provide insight as to whether there is a relationship between mask mandates and car accidents in these two states between April 10, 2020 through December 31, 2020.

```python
import pandas as pd
import datetime
```

```python
# Import csv file containing accidents data & create dataframe

accidentsDF = pd.read_csv("Resources/US_Accidents_Dec20_updated.csv")
accidentsDF.head()
```

```python
# Print list of colums in accidents dataframe

for col in accidentsDF.columns:
    print(col)
```

```python
# Drop irrelevant colums from accidents dataframe

accidentsDF = accidentsDF[["Severity", "Start_Time", "City", "County", "State", "Zipcode"]]
accidentsDF.head()
```

```python
# Drop time stamps from data in "Start-Time" column

accidentsDF['Start_Time'] = pd.to_datetime(accidentsDF['Start_Time']).dt.date
accidentsDF.head()
```

```python
# Print data types for all columns

print(accidentsDF.dtypes)
```

```python
# Convert data in "Start_Time" column to formatted dates

accidentsDF['Start_Time'] = pd.to_datetime(accidentsDF['Start_Time'], format='%Y/%m/%d')
accidentsDF.head()
```

```python
# Rename "Start_Time" column to "Date"

accidentsDF.rename(columns = {'Start_Time':'Date', 'Severity':"Accident Severity"}, inplace = True)
accidentsDF.head()
```

```python
# Drop rows of data prior to 1/1/2020 & select accidents in NY only

accidentsDF = accidentsDF.loc[accidentsDF["Date"] > "2019-12-31"]
NY_accidentsDF = accidentsDF.loc[accidentsDF["State"] == "NY"]
NY_accidentsDF
```

```python
# Select accidents in GA only
GA_accidentsDF = accidentsDF.loc[accidentsDF["State"] == "GA"]
GA_accidentsDF
```

```python
# Import csv containing mask mandate data & create dataframe

mask_mandatesDF = pd.read_csv("Resources/U.S._State_and_Territorial_Public_Mask_Mandates.csv")
mask_mandatesDF.head()
```

```python
# Drop irrelevant columns from mask mandates dataframe

mask_mandatesDF = mask_mandatesDF[["State_Tribe_Territory", "County_Name", "date", "Face_Masks_Required_in_Public"]]
mask_mandatesDF.head()
```

```python
# Print data types  for all columns

print(mask_mandatesDF.dtypes)
```

```python
# Convert data in "date" column to formatted dates

mask_mandatesDF['date'] = pd.to_datetime(mask_mandatesDF['date'], format='%m/%d/%Y')
mask_mandatesDF.head()
```

```python
# Check number of rows 

print(len(accidentsDF.index))
print(len(mask_mandatesDF.index))
```

```python
# Rename columns for readability and consistency

mask_mandatesDF.rename(columns = {'State_Tribe_Territory':'State', 'County_Name':'County', 'date':'Date', 'Face_Masks_Required_in_Public':'Face Masks Req in Public'}, inplace = True)
mask_mandatesDF
```

```python
# Select mask mandates only in NY

NY_mask_mandatesDF = mask_mandatesDF.loc[mask_mandatesDF["State"] == "NY", :]
NY_mask_mandatesDF
```

```python
# Select mask mandates only in GA

GA_mask_mandatesDF = mask_mandatesDF.loc[mask_mandatesDF["State"] == "GA", :]
GA_mask_mandatesDF
```

```python
# Merge accidents dataframe with mask mandate dataframe

NY_mergedDF = pd.merge(NY_accidentsDF,
                 NY_mask_mandatesDF[['Date', 'Face Masks Req in Public']],
                 on='Date')
NY_mergedDF.head()
```

```python
# Reorder columns of merged dataframe 

NY_mergedDF = NY_mergedDF[["Date", "City", "County", "State", "Zipcode", "Accident Severity", "Face Masks Req in Public"]]
NY_mergedDF.head()
```

```python
# Sort merged dataframe by date & reset index 

NY_mergedDF = NY_mergedDF.sort_values('Date').reset_index(drop=True)
NY_mergedDF.head()
```

```python
# Save dataframe to csv file

NY_mergedDF.to_csv('NY_mask_mandates_v_accidents.csv', index=False)
```

```python
# Merge accidents dataframe with mask mandate dataframe

GA_mergedDF = pd.merge(GA_accidentsDF,
                 GA_mask_mandatesDF[['Date', 'Face Masks Req in Public']],
                 on='Date')
GA_mergedDF.head()
```

```python
# Reorder columns of merged dataframe 

GA_mergedDF = GA_mergedDF[["Date", "City", "County", "State", "Zipcode", "Accident Severity", "Face Masks Req in Public"]]
GA_mergedDF.head()
```

```python
# Sort merged dataframe by date & reset index 

GA_mergedDF = GA_mergedDF.sort_values('Date').reset_index(drop=True)
GA_mergedDF.head()
```

```python
# Save dataframe to csv file

GA_mergedDF.to_csv('GA_mask_mandates_v_accidents.csv', index=False)
```
