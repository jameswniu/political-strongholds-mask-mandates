# Final Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os 
print (os.getcwd())
```

```python
NY_mergedDF = pd.read_csv("Resources/NY_mask_mandates_v_accidents.csv")
NY_mergedDF.head()
```

```python
NY_grouped_df = NY_mergedDF.groupby(['Date'])
print(NY_grouped_df)
NY_grouped_df.count().head(10)
```

```python
NY_mask_mandates_passed = NY_grouped_df.count()["Face Masks Req in Public"]
```

```python
plt.figure(figsize=(20,10))
NY_mask_mandates_passed.plot(kind="line")
plt.ylabel("Mask Mandates Passed")
plt.xlabel("Date")
plt.savefig("NY_mask_mandates_per_day.png")
plt.show()
```

```python
x_axis = NY_mergedDF["Date"].unique()
y_axis = NY_mergedDF["Date"].value_counts()

plt.figure(figsize=(20,10))
plt.plot(x_axis, y_axis)
plt.xticks(rotation = 90)
plt.xlim(-0.25, len(x_axis))
plt.tick_params(axis='x', which='major', labelsize=5)
plt.xlabel('Date')
plt.ylabel('Number of Accidents per Day')
plt.savefig("NY_accidents_per_day.png")
plt.show()
```
